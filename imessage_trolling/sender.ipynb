```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

sia = SentimentIntensityAnalyzer() # initialize SIA

# define chem lexicon
chem_lex = {
    'acetal': 5.0,
    'benzene': 4.5,
    'carboxyl': 5.0,
    'alkene': 3.5,
    'amide': 4.0,
    'methyl': 3.4,
    'propyl': 4.0,
    'cyclopentadiene': 6.5,
    'alcohol': 5.0,
    'carbohydrate': 3.9,
    'aldehyde': 9.0,
    'lithium': 4.0,
    'dialkylborane': 10.0,
    'β-elimination': 10.0,
    'aromatic': 5.0,
    'hydride': 10.0,
    'cycloalkane': 10.0,
    'chirality': 10.0,
    'brother': 0.0,
    'hello': 0.0,
    'the': 0.0,
    'hatred': 0.0,
    'sex': 0.0,
    'gun': 0.0,
    'prochirality': 10.0
}

sia.lexicon.update(chem_lex) # update lexicon for SIA

word_vectors = api.load("glove-wiki-gigaword-100")

def get_word_vector(word):
    if word in word_vectors:
        return word_vectors[word]
    else:
        return None
    
def extrapolate_sentiment_score(word, lexicon = chem_lex):
    word_vector = get_word_vector(word)
    if word_vector is None:
        return 0
    similarities = []
    for lex_word, lex_score in lexicon.items():
        lex_vector = get_word_vector(lex_word)
        if lex_vector is not None:
            similarity = cosine_similarity([word_vector], [lex_vector])[0][0]
            similarities.append((similarity, lex_score))
    if not similarities:
        return 0
    total_similarity = sum(similarity for similarity, _ in similarities)
    weighted_score = sum(similarity * score for similarity, score in similarities) / total_similarity
    return weighted_score

def get_sentiment(word):
    # if word in sia.lexicon:
    #     return sia.polarity_scores(word)
    extrapolated_score = extrapolate_sentiment_score(word)
    return {'compound': extrapolated_score}

def makedict(list):
    dick = {}
    for i in list:
        score = extrapolate_sentiment_score(i)
        dick[i] = score
        
    return dick
    
```
