```python
def is_trustworthy(str_input):
    dup_chars = ['а', 'е', 'о', 'р', 'с', 'у', 'х']

    trustworthy = True
    for item in dup_chars:
        if item in str_input:
            trustworthy = False
            break

    return trustworthy
```

```python
def goblinmode(str):
    dup_chars = ['а', 'е', 'о', 'р', 'с', 'у', 'х']

    real_chars = ['a', 'e', 'o', 'p', 'c', 'y', 'x']
    for c in real_chars:
        idx = real_chars.index(c)
        str = str.replace(c, dup_chars[idx])
    
    return str

is_trustworthy(goblinmode("pe"))
```

```python
def factorial(x):
    if x == 1: # This is the base case
        return 1

    else: # This is the recursive case
        print(x)
        return(x * factorial(x-1))
        

print(factorial(4))
```

```python
def bomb(n):
    while n != 0:
        n -= 1
        return bomb(n)
    return 0
n = 100000
fails = 0
for _ in range(n):
    try:
        bomb(n)
        print(f"{n} bomb, {fails} fails")
    except RecursionError:
        n -= 1
        fails += 1
    else:
        break
```
